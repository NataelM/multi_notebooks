# 2.1 Análisis de componentes principales

Un análisis de componentes principales se encarga de explicar la estructura de varianza de un conjunto de combinaciones lineales cuyo objetivo general es reducir la dimensión e interpretación de los datos. Algebraicamente, los componentes principales son una combinación lineal particular de p variables aleatorias $X_1,...,X_p$ que representan un cambio de base obtenido de los nuevos ejes por la rotación del sistema original. Los nuevos ejes representan las direcciones con la máxima variabilidad y proporcionan una descripción simple y parsimoniosa de la estructura de covarianza.

**Note**: Su desarrollo no requiere una suposición normal multivariada, pero los componentes principales derivados de poblaciones normales multivariadas tienen una interpretación útil en términos de elipsoides de densidad constante.. 

Sea el vector aleatorio $\mathbf{x}=(X_1,...,X_p) \in \mathbb{R}^p$ con matriz de covarianzas $\mathbf{\Sigma}$ y valores propios ordenados asociados $\lambda_1 \geq ... \geq \lambda_p \geq 0$  y considere la siguiente combinación lineal dada por los vectores renglon $\mathbf{a_i}$.

$$
\begin{cases}
Y_{1}   & = a_{1 1} X_{1} + \cdots + a_{1 p} X_{p} = \mathbf{a_1 x} \\
\vdots  & = \vdots \\
Y_{p}   & = a_{p 1} X_{1} + \cdots + a_{p p} X_{p} = \mathbf{a_2 x} \\
\end{cases}
\\
\begin{align*}
\left[ \begin{array}{ccccc}
Y_{1} \\
\vdots  \\
Y_{p} \end{array} \right]
& = 
\left[ \begin{array}{ccccc}
a_{1 1} & \cdots & a_{1 p} \\
\vdots  & \ddots & \vdots \\
a_{p 1} & \cdots & a_{p p} \end{array} \right]
\left[ \begin{array}{ccccc}
X_{1} \\
\vdots  \\
X_{p} \end{array} \right]\\
\\
\textbf{y} & =\textbf{Ax} \\
\end{align*}
$$

usando el último resultado, tenemos que el i-ésimo renglon $\mathbf{a_i}$ de $\mathbf{A}$: 

$$
V(Y_i) = \mathbf{a_i \Sigma a_i'}\\
Cov(Y_i,Y_j) = \mathbf{a_i \Sigma a_j'}\\
$$

Sabemos que $Cov(\mathbf{y}) = \mathbf{A \Sigma A'}$ pero las covarianzas de esta expresión pueden aumentar multiplicandose por cualquier constante $\mathbf{A}$, para citar esta indeterminación es necesario restringir a los vectores de coeficientes de longitud 1. Los componentes principales son aquellas combinaciones lineales no correlacionadas $Y_1,...Y_p$ cuyas varianzas son lo más grandes posibles, tales que $Y_i$ es independiente de $Y_j$ o $i \neq j$.

* Primer componente principal: combinación lineal $\mathbf{a_1 x}$ que maximiza $V(\mathbf{a_1 x})$ sujeto a $<\mathbf{a_1,a_1}>$=1
* segunda componente principal: combinación lineal $\mathbf{a_2 x}$ que maximiza $V(\mathbf{a_2 x})$ sujeto a $<\mathbf{a_2,a_2}>$=1 y $<\mathbf{a_1,a_2}>$=0
$$\vdots$$
* La i-ésima componente principal = combinación lineal $\mathbf{a_i x}$ que maximiza $V(\mathbf{a_i x})$ sujeto a $<\mathbf{a_i,a_i}$>=1 y $<\mathbf{a_i,a_k}>$=0 para $k < i$

Dado que la matriz de covarianza $\mathbf{\Sigma}$ es ortogonalmente diagonalizable, $\mathbf{\Sigma}=\mathbf{\Gamma \Lambda \Gamma'}$, si  $\lambda_k \mathbf{\gamma_k}$ es un vector propio asociado a $\mathbf{\Sigma}$, se satisfacen las restricciones de los componentes principales:

1. $<\mathbf{\gamma_i',\gamma_i}$>=1
1. $<\mathbf{\gamma_i',\gamma_0}$>=0 para $k < i$

Por demostrar que la varianza se puede ver como funcion de $\mathbf{a}$ y los vectores propios maximizan la varianzas. Considera las restricciones que $\mathbf{a}$ es claro que $\mathbf{a' \Sigma a}$ y $\mathbf{\frac{a' \Sigma a}{a' a}}$ tienen el mismo maximo en $\mathbf{\frac{a' \Sigma a}{a' a}}$ elimina el problema de la constante. Considere la siguiente transformación $\mathbf{b}=\mathbf{\Gamma' a}$ Se tiene:

$$
\begin{align*}
\mathbf{\frac{a' \Sigma a}{a' a}} &= \mathbf{\frac{a' \Gamma \Lambda \Gamma' a}{a' a}}\\
&= \mathbf{\frac{b' \Lambda b}{b' b}} \\
\mathbf{\frac{a' \Sigma a}{aa'}} &= \frac{\sum_{k=1}^{p} \lambda_k b_k^2}{\sum_{k=1}^{p}b_k^2}\\
\end{align*}
$$

Dado que  $\lambda_p \leq \lambda_{p-1} \leq \dots \leq \lambda_i \leq \dots \leq \lambda_1$ para el i-ésimo valor propio $\lambda_i$ la varianza como función de la función $\mathbf{a}$.

$$
\begin{cases}
\lambda_i b_1^2 \leq \lambda_i b_1^2  \\
\lambda_{i-1} b_1^2 \leq \lambda_i b_1^2  \\
\vdots  \\
\lambda_p b_p^2 \leq \lambda_1 b_p^2  \\
\end{cases}
$$


$$
\begin{align*}
\lambda_i b_1^2 + \lambda_{i-1} b_2^2 + \dots +\lambda_p b_p^2 &\leq \lambda_i b_1^2 + \lambda_i b_2^2 + \dots +\lambda_i b_p^2\\
\frac{\sum_{k=i}^{p} \lambda_k b_k^2}{\sum_{k=i}^{p}b_k^2} &\leq \frac{\sum_{k=i}^{p} \lambda_i b_k^2}{\sum_{k=i}^{p}b_k^2}\\
\frac{\sum_{k=i}^{p} \lambda_k b_k^2}{\sum_{k=i}^{p}b_k^2} &\leq \lambda_i \frac{\sum_{k=i}^{p} b_k^2}{\sum_{k=i}^{p}b_k^2}\\
\frac{\sum_{k=i}^{p} \lambda_k b_k^2}{\sum_{k=i}^{p}b_k^2} &\leq \lambda_i \\
\end{align*}
$$

En particular para $\lambda_1$ tenemos que cualquier varianza  para cualquier combinacion lineal con coeficientes dados por $\mathbf{a}$ se sigue del primer valor propio.

$$
\begin{align*}
\frac{\sum_{k=1}^{p} \lambda_k b_k^2}{\sum_{k=1}^{p}b_k^2}&\leq \lambda_1 \\
\mathbf{\frac{a' \Sigma a}{a' a}}&\leq \lambda_1 \\
\end{align*}
$$

Si algún valor de $\mathbf{x}$ la función alcanza el máximo, entonces este punto es un punto critico y maximiza la función, entonces reemplazando el primer valor propio $\gamma_1$:

$$
\begin{align*}
\mathbf{\frac{a \Sigma a'}{a' a}} &= \mathbf{\frac{\gamma_i' \Sigma \gamma_i}{\gamma_i' \gamma_i}} \\
&= \mathbf{\gamma_i' \Sigma \gamma_i} \\
&= \mathbf{\gamma_i'} \lambda_i \mathbf{\gamma_i} \\
&= \lambda_i \\
\mathbf{\gamma_i' \Sigma \gamma_i} &= \lambda_i \\
V(\mathbf{\gamma_i' x}) &= \lambda_i \\
\end{align*}
$$

si tomamos $\mathbf{a}=\mathbf{\gamma_i'}$ esta transformación alcanza la varianza maxima para $Y_i = \mathbf{\gamma_i' x}$, entonces $\mathbf{\gamma_i'}$ es su i-ésima componente principal, porque estas son las combinaciones lineales que maximizan las varianzas y producen nuevas variables independientes. 

![pca gif](static/pca2gif.gif)

Entonces, la transformación de componentes principales de $\mathbf{x}$ con matriz de covarianzas dada por $Cov(\mathbf{x})=\mathbf{\Sigma}=\mathbf{\Gamma \Lambda \Gamma'}$  es:


$$
\begin{cases}
\begin{align*}
PCA_1 & = \gamma_{1 1} X_{1} + \cdots + \gamma_{1 p} X_{p} = \mathbf{\gamma_1' x} \\
\vdots & = \vdots \\
PCA_p & = \gamma_{p 1} X_{1} + \cdots + \gamma_{p p} X_{p} = \mathbf{\gamma_p' x} \\
\end{align*}
\end{cases}
$$


$$
\begin{align*}
\begin{bmatrix}
    PCA_1 \\
     \vdots\\
    PCA_p \\
\end{bmatrix}&=
\begin{bmatrix}
    \text{-} & \mathbf{\gamma_1} & \text{-} \\
    & \vdots &\\
    \text{-} & \mathbf{\gamma_p} & \text{-} \\
\end{bmatrix}
\begin{bmatrix}
    X_1 \\
     \vdots\\
    X_p \\
\end{bmatrix}\\
PCA(\mathbf{x}) &= \mathbf{\Gamma' x}  \\
\end{align*}
$$


Si tenemos la matriz de datos $\mathbf{X}_{n \times p}$ con $\mathbf{S} = \mathbf{\Gamma \Lambda \Gamma'}$ la transformacion de PCA transformations para esta matriz está dada por:

$$
PCA(\mathbf{X}_{n \times p}) = \mathbf{X}_{n \times p}\mathbf{\Gamma}
$$

**Razones de variabilidad explicada:**

Del análisis espectral tenemos 

$$
\begin{align*}
\Sigma&=\sum_{k=1}^p\lambda_k\mathbf{\gamma_k\gamma_k'}\\
tr[ \Sigma ] &= tr \left[ \sum_{k=1}^p\lambda_k\mathbf{\gamma_k\gamma_k'} \right]\\
&= \sum_{k=1}^p \lambda_k tr \left[ \mathbf{\gamma_k\gamma_k'} \right]\\
&= \sum_{k=1}^p \lambda_k tr \left[ \mathbf{\gamma_k'\gamma_k} \right]\\
tr[ \Sigma ] &= \sum_{k=1}^p \lambda_k \\
\end{align*}
$$

El porcentaje de variabilidad que contribuye al i-ésimo componente está dado por

$$\frac{\lambda_i}{tr[\mathbf{\Sigma}]}=\frac{\lambda_i}{\lambda_1+\dots+\lambda_p}$$

El porcentaje acumulado de variabilidad hasta el i-ésimo componenete es:

$$\frac{\sum_{k=1}^{i}\lambda_k}{tr[\mathbf{\Sigma}]}=\frac{\lambda_1+\lambda_i}{\lambda_1+\dots+\lambda_p}$$


El número de componentes principales se toma de modo que la variabilidad explicada acumulada se acerque a un valor especificado por el científico, por ejemplo 80%. Por otro lado, si la representación se estabiliza a partir de un valor propio determinado, entonces aumentar la dimensión apenas proporciona una variabilidad más explicada.

# Biplot
---

Un biplot es una generalización del diagrama de dispersión simple de dos variables. Un biplot permite mostrar gráficamente información sobre muestras y variables de una matriz de datos. Las muestras se muestran como puntos, mientras que las variables se muestran como vectores. El prefijo "bi" se refiere a la superposición, en la misma parcela de variables y observaciones.

Un biplot para una matriz $\mathbf{X_{n \times p}}$ es un gráfico representado por p vectores (marcadores) $v_1,\dots,v_p$ para los n renglones $u_1,\dots, u_n$ de $\mathbf{X}$ tal que el elemento $x_{ij} \approx \mathbf{v_i u_j'}$ lo mejor posible.

$$\mathbf{X \approx UV'}$$

Para el espacio euclideano $\mathbb{R^2}$ si la matriz $\mathbf{X}$ tiene rango 2 las aproximaciones serán exactas en otro caso serán una aproximación. Considere la transformacion de PCA:

$$
\begin{align*}
\mathbf{\Gamma' x}&=\mathbf{y}\\
\mathbf{x} &=\mathbf{\Gamma y}\\
\end{align*}
\\
\begin{cases}
X_{1}   & = \gamma_{1 1} Y_{1} + \cdots + \gamma_{p 1} Y_{p}\\
\vdots  & = \vdots \\
X_{p}   & = \gamma_{1 p} Y_{1} + \cdots + \gamma_{p p} Y_{p}\\
\end{cases}
$$

Tomando el primer componente principal.

$$
\begin{cases}
\begin{align*}
X_{1}   & \approx \gamma_{1 1} Y_{1} + \cdots + \gamma_{i 1} Y_{i}\\
\vdots  & \approx \vdots \\
X_{p}   & \approx \gamma_{1 p} Y_{1} + \cdots + \gamma_{i p} Y_{i}\\
\end{align*}
\end{cases}
$$

Los marcadores serán las primeras columnas de las variables transformadas al nuevo espacio y los vectores las primeras columnas de la matriz de cambio de base. Esta aproximación depende del número de componentes principales tomadas (normalmente 2), la resolución del biplot viene dada por el porcentaje de variabilidad acumulado hasta la i-ésima componente.

```python
features = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
keys = ['Species']
irisPath = 'DataSets/Iris.csv'
irisPD = pd.read_csv(irisPath)
irisPD
```

Estandarizando la data Para PCA, el no estándarizar generaría un problema pues aquellas variables cuyos valores sean más altos, cargaran más varianza al PCA

```python
from sklearn.preprocessing import StandardScaler

X = irisPD.loc[ : , features].values
Y = irisPD.loc[ : , keys].values

# matriz de datos centrada
centeredIrisX = StandardScaler(with_mean=True,with_std=True).fit_transform(X)
pd.DataFrame(centeredIrisX)
```

```python
from sklearn.decomposition import PCA

k=4
pcaIrisCols = ['pc'+str(i+1) for i in range(0,k)]
pcaIrisModel = PCA(n_components=k).fit(centeredIrisX)

# este tranform es la matriz de datos multiploicada por la matriz de vectores propios(gamma) (cambio de base)
transformedIrisDataPD = pd.DataFrame(data=pcaIrisModel.transform(centeredIrisX),columns=pcaIrisCols)
pcaIrisPD = pd.concat([irisPD,transformedIrisDataPD],axis=1)
pcaIrisPD[:5]
```

```python
#matriz Gamma = Matriz de cargas = matriz de transformación = matriz de vectores propios de la matriz de var y cov
pcaIrisMatrix = pd.DataFrame(pcaIrisModel.components_.transpose(),columns=pcaIrisCols)
pcaIrisMatrix['feature'] = features

print('PCA Transformation Matrix:\n')
display(pcaIrisMatrix)
print('PCA Eigenvalues:\n',pcaIrisModel.explained_variance_)
print('PCA Explainded variance:\n',pcaIrisModel.explained_variance_/sum(pcaIrisModel.explained_variance_))
```

```python
kw = 12
labelCol = 'label'
whiteWinnePath = 'DataSets/winequalityWhite.csv'
redWinnePath = 'DataSets/winequalityRed.csv'
```

```python
whiteWinePD = pd.read_csv(whiteWinnePath)
redWinePD = pd.read_csv(redWinnePath)
features = list(redWinePD.columns)

#columnas de etiqueta
whiteWinePD['type'] = 0
redWinePD['type'] = 1
display(whiteWinePD)
display(redWinePD)

#concatenado de las tablas de los vinos
wineX = pd.concat([whiteWinePD,redWinePD])
wineX
```

### **Sin estandarizar las variables**

```python
kw=2
pcaWineCols = ['pc'+str(i+1) for i in range(0,kw)]
features = ['fixedAcidity', 'volatileAcidity', 'citricAcid', 'residualSugar','chlorides', 
            'freeSulfurDioxide', 'totalSulfurdioxide', 'density', 'pH','sulphates', 'alcohol', 'quality']
centeredWineX = wineX[features]
centeredWineX
```

```python
pcaWineModel = PCA(n_components=kw).fit(centeredWineX)

#matriz de transformación o cargas es el .components_
pcaWineMatrix = pd.DataFrame(pcaWineModel.components_.transpose(),columns=pcaWineCols[:kw])
pcaWineMatrix['cols'] = features
display(pcaWineMatrix)

pcaRedWineDataPD = pd.DataFrame(pcaWineModel.transform(wineX[wineX['type']==1][features]),columns=pcaWineCols[:kw])
pcaWhiteWineDataPD = pd.DataFrame(pcaWineModel.transform(wineX[wineX['type']==0][features]),columns=pcaWineCols[:kw])

display(pcaRedWineDataPD)
display(pcaWhiteWineDataPD)
```

Cómo se puede observar en la primera tabla de lo que aporta cada variable a las dos primeras componentes, podemos que:
freeSulfurDioxide
totalSulfurdioxide

se llevan toda la varianza, Esto podría ser debido a la escala de dichas variables


```python
print('PCA Transformation Matrix:\n')
display(pcaWineMatrix)
plt.figure(figsize=(5,5))
screePlot(plt,pcaWineModel.explained_variance_ratio_)
```

```python
plt.figure(figsize=(15,15))
scatterPlot(plt,pcaWhiteWineDataPD,'pc1','pc2','Blue',color='Black',alpha=0.5)
scatterPlot(plt,pcaRedWineDataPD,'pc1','pc2','Red',color='Black',alpha=0.5)
plt.show()
```

Cuando las componentes tienen una forma picuda (conica) es para pensar que algo anda mal con los componentes. Una inspección rápida revela que hay algunas columnas con una varianza alta, estas columnas están absorbiendo la covarianza de los componentes, esto sucede porque todas las columnas no están en la misma escala métrica, estandarizar las columnas podría ser una buena transformación para evitar que algunas columnas absorban toda la varianza explicada

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import QuantileTransformer

#scaler = RobustScaler().fit(wineX[features])
#scaler = Normalizer().fit(wineX[features])
#scaler = QuantileTransformer(n_quantiles=10).fit(wineX[features])
scaler = StandardScaler(with_mean=True,with_std=True).fit(wineX[features])
stdWineX = pd.DataFrame(scaler.transform(wineX[features]),columns=features)
stdWineX.plot.box(figsize=(17,10))
stdWineX
```
