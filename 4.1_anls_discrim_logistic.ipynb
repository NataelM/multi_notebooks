# Regresión Logistica

---

Considere dos poblaciones $\Omega_1$, $\Omega_2$ con vectores aleatorios de caracteristicas asociados $\mathbf{x_1},\mathbf{x_2} \in \mathbb{R}^p$, un discriminante es una función $D:\mathbb{R}^p \rightarrow \mathbb{R}$ tal que si $D(\mathbf{x}) \geq t$ para algun umbral $t \in \mathbb{R}$ entonces se asigna $\mathbf{x}$ a $\Omega_1$ en otro caso $\mathbf{x}$ a $\Omega_2$. D es un discriminante lineal si D tiene la forma:

$$D(\mathbf{x})=\beta_0+\beta_1 x_1 + \dots + \beta_p x_p = \beta_0+\mathbf{x' \beta}$$

Se supone que existe una linea recta que separa de forma lineal a los conjuntos de datos.

### Modelo Logistico
___

Considere el siguiente vector aleatorio $(Y,X_1,\dots,X_p) \in {0,1} \times \mathbb{R}^p$, donde $Y$ es una variable indicadora y esta toma el valor de 1 si el vector aleatorio pertence a la población $\Omega_1$ y 0 si pertenece a  $\Omega_2$, entonces tiene sentido que la probabilidad de pertenecer a $\Omega_1$ (objetive population) debe ser parametrizada por un discriminante D, donde D es una combinación lineal de caracteristicas ponderadas:

$$P[Y=y|X_1=x_1,\dots,X_p=x_p]=p(D)=p(\mathbf{x};\beta_0,\mathbf{\beta})$$


El Modelo logistico establece que el incremente en la probabilidad de pertenecer a la población $\Omega_1$ con relación añ discriminante D es directamente proporcional a la probabilidad de pertenecer a $\Omega_1$ dada D y disminuye a medidad que se acerca a 1. 

$$ \frac{p'}{p}=1-p$$

El objetivo es intentar resolver la ecuación diferencial para tener a $p$ como función d elos valores de D.

$$
\begin{align*}
\frac{p'}{p} &= 1-p\\
p' &= p(1-p)\\
\frac{dp}{dD} &= p(1-p)\\
\frac{dp}{p(1-p)}&=dD\\
\int \frac{dp}{p(1-p)}&=\int dD\\
\int \frac{dp}{p-p^2}&=D\\
\end{align*}
$$

Tenga en cuenta que la constante de la integral en el lado derecho podría ser absorbida por el intercepto de D, y el problema se reduce a  resolver la integral del lado izquierdo. 
Usando fracciones parciales:

$$
\begin{align*}
\frac{1}{p(1-p)} &= \frac{A}{p}+\frac{B}{1-p} \\
1 &= \frac{A}{p}(p(1-p))+\frac{B}{1-p}(p(1-p)) \\
1 &= A(1-p)+Bp\\
& \left \{ \begin{matrix} - B &= -1
\\ A+B &= 0 \end{matrix}\right.\\
& A = 1, B = -1 \\
\implies \frac{1}{p-p^2} &= \frac{1}{p} - \frac{1}{1-p} \\
\end{align*}
$$

Volviendo a la solución de la ecuación diferencial:

$$
\begin{align*}
\int \frac{dp}{p-p^2} &= D \\
\int \frac{1}{p}dp - \int \frac{1}{1-p}dp &= D\\
ln(p)-ln(1-p) &= D\\
ln\left(\frac{p}{1-p}\right) &= D  \dots (1)\\
\frac{p}{1-p} &= e^D\\
p &= (1-p) e^D\\
p &= e^D-pe^D\\
p + pe^D &= e^D\\
p(1+e^D) &= e^D\\
p &= \frac{e^D}{1+e^D}\\
p &= \frac{1}{1+e^D}\frac{1}{e^{-D}}\\
p &= \frac{1}{1+e^{-D}}\\
p(\mathbf{x}) &= \frac{1}{1+e^{-(\beta_0+\mathbf{x' \beta}})} ... (2)\\
\end{align*}
$$

De **ecuación (1)** el lado izquierdo se conoce como **logit transformation**, y es una forma de estimar una proporción de pertenecer o no a la población objetivo.. 

$$ ln\left(\frac{p(\mathbf{x})}{1-p(\mathbf{x})}\right) = \beta_0+\beta_1 x_1 + \dots \beta_p x_p$$

La transformación da una biyección de la escala de probabilidad $\left[0,1\right]$ a la escala logística $\left[−\infty,\infty\right]$. 

**La ecuación (2)** dada la proba de pertenecer a $\Omega_1$ en función de las características de los pesos y de las características de $\Omega_1$, el argumento del logaritmo $\frac{p}{1-p}$ se llama razón de probabilidades. Esta ecuación es llamda la **Ecuación logistica o sigmoide**. 

$$p(\mathbf{x}) = \frac{1}{1+e^{-(\beta_0+\beta_1 x_1 + \dots  + \beta_p x_p)}}$$

### Regresión Logistica
___

Dado que la variable indicadora sigue una distribución de Bernoulli, la función de Verosimilitud asociada a $p(\mathbf{x})$ Para la matriz de datos $\mathbf{X} \in M_{n \times p+1}(\mathbb{R})$, donde cada fila son observaciones independientes sin estar idénticamente distribuidas en general:

$$
\mathcal{L}(\beta_0,\beta_1,...,\beta_p) = \prod_{i=1}^{n} p(\mathbf{x}_i)^{y_i}(1-p(\mathbf{x}_i))^{1-y_i}
$$

LA función de log-berosimilitud es:

$$
\begin{align*}
ln \mathcal{L}(\beta_0,\beta_1,...,\beta_p) &= \sum_{i=1}^{n} y_i ln \left(p(\mathbf{x}_i)\right)+\sum_{i=1}^{n} (1-y_i)ln \left(1-p(\mathbf{x}_i)\right) \\
\end{align*}
$$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
from sklearn.decomposition import PCA
from IPython.display import display, HTML
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score

pd.set_option('max_colwidth',100)
warnings.filterwarnings('ignore')
plt.style.use('ggplot')
```

```python
def scatterPlot(plt,datasetPD,col_1,col_2,colorDots='darkblue',label=None,shift=0.05,**kwargs):
    dataPD = datasetPD[[col_1,col_2]]
    plt.scatter(dataPD[col_1],dataPD[col_2],color=colorDots)
    if(label!=None):
        points = datasetPD[[col_1,col_2,label]].values
        for point in points:
            plt.text(point[0]+shift,point[1]+shift,str(point[2]))
    plt.axvline(**kwargs)
    plt.axhline(**kwargs)
```

```python
def logisticSurface(coeficients,intercept,vector):
    logitTransformation = np.dot(coeficients,vector)+intercept
    return 1/(1+math.exp(-1*logitTransformation))
```

```python
def plot_confusion_matrix(plt,y_true, y_pred, figsize=(10,10), font_size=16):
    cm = confusion_matrix(y_true, y_pred, labels=np.unique(y_true))
    cm_sum = np.sum(cm, axis=1, keepdims=True)
    cm_perc = cm / cm_sum.astype(float) * 100
    annot = np.empty_like(cm).astype(str)
    nrows, ncols = cm.shape
    for i in range(nrows):
        for j in range(ncols):
            c = cm[i, j]
            p = cm_perc[i, j]
            if i == j:
                s = cm_sum[i]
                annot[i, j] = '%.1f%%\n%d/%d' % (p, c, s)
            elif c == 0:
                annot[i, j] = ''
            else:
                annot[i, j] = '%.1f%%\n%d' % (p, c)
    cm = pd.DataFrame(cm, index=np.unique(y_true), columns=np.unique(y_true))
    cm.index.name = 'Actual'
    cm.columns.name = 'Predicted'
    fig, ax = plt.subplots(figsize=figsize)
    sns.heatmap(cm, cmap= "Blues", annot=annot, fmt='', ax=ax, annot_kws={"size": font_size})
```

```python
df = pd.read_csv('DataSets/loans_.csv')
display(df)
plt.figure(figsize=(10,10))
scatterPlot(plt,df[df['loan_status']==0],'fico_score','income','Blue',color='Black',alpha=0.6)
scatterPlot(plt,df[df['loan_status']==1],'fico_score','income','Red',color='Black',alpha=0.6)
```

```python
# se logaritmiza el fico score y los ingresos
df['fico_score'] = np.log(df['fico_score'])
df['income'] = np.log(df['income'])

plt.figure(figsize=(10,10))
scatterPlot(plt,df[df['loan_status']==0],'fico_score','income','Blue',color='Black',alpha=0.6)
scatterPlot(plt,df[df['loan_status']==1],'fico_score','income','Red',color='Black',alpha=0.6)
```

```python
import statsmodels.api as sm

X = sm.add_constant(df[['fico_score','income']])
Y = df['loan_status']

display(X)

logistic_model = sm.Logit(Y, X).fit(disp = 0) #disp = 0 sólo es la inicialización del algoritmo
display(logistic_model.summary())
```

Un p_value de cero indica que esa $\beta_i$ no es cero con la proba del p_value.

Además una $\beta_i = 0$ se puede interpretar como que esa variable NO ayuda a separar a los datos.
Por otro lado el coeficiente de las $\beta_i$ se pueden interpretar como auqel nivel de separación que aportan para separar las clases.

```python
intercept = logistic_model.params[0]
coeficients = list(logistic_model.params[1:]) 
```

```python
from mpl_toolkits.mplot3d import Axes3D
import math

positives_df = df[df['loan_status']==1]
negatives_df = df[df['loan_status']==0]

positivesX1 = list(positives_df['fico_score'].values)
positivesX2 = list(positives_df['income'].values)
negativesX1 = list(negatives_df['fico_score'].values)
negativesX2 = list(negatives_df['income'].values)

plt.style.use('seaborn-whitegrid')
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(positivesX1, positivesX2, 0.5, color="Red", marker='o')
ax.scatter(negativesX1, negativesX2, 0.5, color="Blue", marker='o')

ax.set_xlabel('fico_score')
ax.set_ylabel('income')
ax.set_zlabel('$\hat{p}(\mathbf{x})$')

x1 = np.linspace(df['fico_score'].min(),df['fico_score'].max(),100)
x2 = np.linspace(df['income'].min(),df['income'].max(),100)
X1, X2 = np.meshgrid(x1,x2)
Z = np.array([[logisticSurface(coeficients,intercept,[x1,x2]) for x1 in x1] for x2 in x2])
ax.plot_surface(X1,X2,Z,cmap='coolwarm',alpha=0.6,linewidth=0)
ax.view_init(45,65)
plt.show()
```

```python
sns.histplot(df[df['loan_status']==0]['score'],color='blue',bins=50);
sns.histplot(df[df['loan_status']==1]['score'],color='red',bins=50);
```

Dadas las distribuciones podemos decir que existen diferencias entre los que impagan y los que sí pagan

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import roc_curve, roc_auc_score

                                        
fpr, tpr, test_thresholds  = roc_curve(df['loan_status'], #valores reales
                                       df['score']) #valores predichos (en probabilidad)

plt.figure(figsize=(10,10))
plt.plot(fpr, tpr, linestyle='-', color='blue', label='Training ROC')
plt.plot([0,1],[0,1], color='black', label='Randomness Base Line')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

#calcular el area debajo de la curva
print('AUC:',roc_auc_score(df['loan_status'], #reales
                           df['score']))  #predichos (proba)


print('GINI:',2*roc_auc_score(df['loan_status'],df['score'])-1)
```

Para la curva ROC: valores menores a 0.5 indica que no podemos separar bien las clases de las etiquetas
y mientras el area de la curva ROC se acerque a 1 nos indica que nuestro modelo puede separar mejor las clases.

Lo anterior se traduce que para la etiqueta 1 tenemos altas probabilidades de etiquetarlo de manera correcta, es decir es una probailidad que da cierta (confianza de que un elemento de la clase positiva tenga mayor proba que uno de la clase negativa) (es decir, qué tan bueno es el modelo scoreando etiquetas de cero y uno)

El Gini de clasificacion mide el doble del area entre la identidad y la curva roc

**En clasificación es importante decidir a partir de qué punto (umbral-proba (de prediccion) ) podemos considerar la clasificacion de 1, a esto le llamamos _punto de corte_**

```python
df.sort_values('score',ascending=True,inplace=True)

# para poder tener un buen nivel de corte de una manera no arbitraria podemos hacer lo siguiente 
df['profit'] = np.where(df['loan_status']==0 , # de los que sí pagan
                        df.amount*1.25, #asigna la ganancia de ese prestamo 
                        -df.amount) #si no pagan asigna perdida del prestamo por ese monto

# ganancia acumulada por presatamo (la tabla ya debe de estar ordenada por score)
df['cumulated_profit'] = df['profit'].cumsum()

#entonces estableceremos el umbral optimo como aquel que maximiza la ganancia acumuladad (cumulated_profit)
threshold_row = df[df['cumulated_profit'] == df['cumulated_profit'].max()]# aquí está el máximo
threshold_row #punto de corte
```

El máximo lo podemos ver en un score de 0.48, y de aquí podemos decir que no aceptaremos a personas con un score mayor a 0.48 pues mayor a ese score la gente ya no paga, entonces ya no doy el credito

```python
threshold_cutoff = threshold_row['score'].values[0] # punto de corto 

#genero un target_estimado según el threshold
df['target_est'] = np.where(df['score']>=threshold_cutoff,
                            1, # si el score es mayor  al threshold (no doy el credito)
                            0) # su el score es menor al threshold (Sí doy el crédito)
df
```

Métricas estáticas de clasificación

```python
precision = precision_score(df['loan_status'],df['target_est'])
recall = recall_score(df['loan_status'],df['target_est'])
f1 = f1_score(df['loan_status'],df['target_est'])
accuracy = accuracy_score(df['loan_status'],df['target_est'])

print('HEAD:')
display(df.head(5))
print('TAIL:')
display(df.tail(5))
print('OPTIMAL THRESHOLD CUTOFF SCORE VALUE:')
display(threshold_row)

plt.figure(figsize=(10,10))
plt.axvline(x=threshold_cutoff,label='Optimal profit')
plt.plot(df['score'].values,df['cumulated_profit'].values, linestyle='-', color='Black', label='Cumulated profit')
plt.plot([1,0],[1,0], linestyle='-', color='Gray', label='Base Line')

plt.xlabel('Score')
plt.ylabel('Cumulated profit')
plt.legend()
plt.show()

print('PRECISION:',precision)
print('RECALL:',recall)
print('F1:',f1)
print('ACURACY:',accuracy)
print('DENIED OPERATIONS:',len(df[df['score']>=threshold_cutoff]))
print('DENIED PERCENTAGE:',len(df[df['score']>=threshold_cutoff])/len(df))

plot_confusion_matrix(plt,df['loan_status'],df['target_est'],figsize=(10,10),font_size=20)

odds = np.log(threshold_cutoff/(1-threshold_cutoff))
# odds = b0+b1x1+b2x2

#recta que separa a azules de rojos 
beta_0 = logistic_model.params[0]
beta_1 = logistic_model.params[1]
beta_2 = logistic_model.params[2]


x1 = np.linspace(df['fico_score'].min(), df['fico_score'].max(), 1000)
#despejo x_2 para encontrar la recta que separa a los datos
x2 = (odds-beta_0-beta_1*x1)*(1/beta_2)

plt.figure(figsize=(10,10))
scatterPlot(plt,positives_df,'fico_score','income','Blue',color='Black',alpha=0.6)
scatterPlot(plt,negatives_df,'fico_score','income','Red',color='Black',alpha=0.6)
plt.plot(x1, x2,color='Black',alpha=0.6)
```
