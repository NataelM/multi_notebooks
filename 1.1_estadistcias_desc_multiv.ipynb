# Data sets Multivariados

#### Matriz de datos

Supon que tienes n observaciones de un vector aleatrio $\vec{X}$ (distribución de la porblacion), tal que cada vector tiene p variables explicativas. ENtonces el conjunto de observaciones $\{\vec{X}_i\}_{i=1}^n=\{(X_1,...,X_p)_i\}_{i=1}^n$ puede ser representada como una matriz, llamada matriz de datos $\textbf{X}_{n \times p}$, los renglones de la magtriz representan el índice de las observaciones y las columnas tepresentan las variables explicativas.

$$\textbf{X}_{n \times p}=
\left[ \begin{array}{ccccc}
x_{1 1} & \cdots & x_{1 j} & \cdots & x_{1 p} \\
\vdots  & \ddots & \vdots  & \ddots & \vdots \\
x_{i 1} & \cdots & x_{i j} & \cdots & x_{i p} \\
\vdots  & \ddots & \vdots  & \ddots & \vdots \\
x_{n 1} & \cdots & x_{n j} & \cdots & x_{n p} \end{array} \right]
$$

**Notación:**
* $\textbf{x}_i$ indica el i-ésimo renglon de $\textbf{X}$, no obstante podemos operar de manera columnar.
* $X_j$ indica la j-ésima columna de  $\textbf{X}$

**Example 1.1 - 4 [Centering data sets]**

Aunque las expresiones matriciales ayudan a obtener resultados teóricos útiles, su implementación es demasiado costosa desde el punto de vista computacional, por lo que preferimos utilizar métodos basados en álgebra relacional para obtener estadísticas descriptivas.

```python

##################### Matriz de datos centrado vía algebra lineal ########################################
import numpy as np

n = len(irisPD) #largo de la tabla
nn_1 = []
for i in range(0,n):
    nn_1.append([1]*n)
    
nn_1 = np.matrix(nn_1) # MAtriz de puros unos cuadrada de nxn

# la identidad en "n" menos la matriz
H = np.matrix(np.identity(n)) - (1/n)*nn_1 # matriz identidad menos 1/n multiplicada por la matriz de puros unos  
X = np.matrix(irisPD[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values) #extrayendo matriz de datos 
X_centered = np.dot(H,X) #matriz de centrado = {matriz id - [(1/n)*matriz de unos ]} X matriz de datos 
 
centered_iris_df = pd.concat([pd.DataFrame(X_centered, columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']),
                              irisPD[['Species']]],
                             axis=1)
centered_iris_df
```

***Matriz de datos centrado restando unicamente la media de cada columna***

```python
irisMeansPD = irisPD.mean()

#vector de medias
irisMeansPD#.to_dict()['SepalLengthCm']
```

```python
def getCenteredDataMatrix(dataPD, columns, keys):
    '''
    dataPD: dataframe 
    columns: columnas a centrar
    keys: mantener los "id's"
    
    regresa un dataframe con los datos centrados 
    '''
    dataframePD = dataPD.copy(deep=True) # copia del data frame
    meanVector = dataPD.mean().to_dict() # vector de medias
    for col in columns: # for para cada columan 
        #se contruye una columna con los datos centrados 
        dataframePD['centered_'+col] = dataframePD[col] - meanVector[col]
    return dataframePD[keys + ['centered_' + col for col in columns]]
```

```python
columns = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'] #columnas a centrar
keys = ['Species'] # columna del id
centeredIrisPD = getCenteredDataMatrix(irisPD,columns,keys)
centeredIrisPD
```

También este analisis se suele hacer de manera marginal como se muestra en las densidades de arriba, y podemos ver que el ancho del sepalo podría separar mejor 

```python
def overlapedHitograms(plt,datasetPD,columns,target,labels=None,sample=1.0,bins=10,size=(17,8),alpha=0.7):
    '''
    función que pinta los histogramas
    
    '''
    plt.style.use('ggplot')
    fig = plt.figure(figsize=size)
    if labels == None:
        labels = [0,1]
    negativesPD = datasetPD[datasetPD[target]==labels[0]] #versicolor
    positivesPD = datasetPD[datasetPD[target]==labels[1]] #virginica
    for i, column in enumerate(columns):
        negativeValues = negativesPD[column].values.transpose()
        positiveValues = positivesPD[column].values.transpose()
        ax = fig.add_subplot(2, 2, i+1,)
        ax.set_title('discriminant power '+column)
        ax.hist(negativeValues,bins=bins,alpha=alpha)
        ax.hist(positiveValues,bins=bins,alpha=alpha)
```

```python
centeredFeatures = [col for col in centeredIrisPD.columns if col not in keys]
labels = ['Iris-versicolor','Iris-virginica']
overlapedHitograms(plt,centeredIrisPD,centeredFeatures,keys[0],labels)
```
