#### Distancias y Disimilaridades
Dados dos objetos $x$ y $y$ en cualquier espacio $\Omega$, la distancia (disimilaridad) entre esos objetos es una funci√≥n $d$ es una medida numerica de que tan diferentes son 2 objetos: 

sea la disimilaridad $d(x,y): \Omega \times \Omega \rightarrow \mathbb{R^+} \cup \{0\}$ tal que:
* $d(x,y) \ge 0$
* $d(x,y)=0$ si y solo s√≠ $x=y$
* $d(x,y)=d(y,x)$
* $d(x,y) \leq d(x,z)+d(z,y)$

Una distancia que satidface esas propiedades es llamada metrica. dados $\mathbf{x,y} \in \mathbb{R}^p$ las siguientes funciones son distancias:

* Distancia Euclideana:

$d_E(\mathbf{x,y}) = \sqrt{(x_1-y_1)^2+ \dots (x_p-y_p)^2} = \sqrt{\mathbf{(x-y)'(x-y)}}$

* Distancia de Minkowsky:

$d(\mathbf{x,y}) = \left[ \sum_{i=1}^p \left|x_i - y_i \right| \right]^{\frac{1}{\lambda}}$

* Distancia de Mahalanobis:

$d(\mathbf{x,y}) = \sqrt{\mathbf{(x-y)'S^{-1}(x-y)}}$

#### Similaridades

Una medidad de similaridad es una funcion $s$ de que tan parecidos son dos objetos. 

Sea la similaridad $s(x,y): \Omega \times \Omega \rightarrow \mathbb{R}$ tal que

* $s(x,y) \leq 1$
* $s(x,y)=1$ if and only if $x=y$
* $s(x,y)=s(y,x)$

Las siguientes funciones son similaridades:

* Similaridad del coseno: dada por $\mathbf{x,y} \in \mathbb{R}$

$s(\mathbf{x,y}) = \mathbf{\frac{<x,y>}{||x||||y||}}$

* Las correlaciones: Dados X,Y variables aleatorias

$s(X,Y) =  \frac{Cov(X,Y)}{\sqrt{Var(X) Var(Y)}}$

* EL √≠ndice de Jaccard: Dados X,Y conjuntos finitos

$s(X,Y) = \frac{|X \cap Y|}{|X \cup Y|}$

### Metrica de escalamiento multidimensional MDS
___ 

El objetivo de una metrica de escalamiento multidimensional (MDS) es mapear un conjunto finito $\Omega$ a un espacio m√©trico euclidiano $\mathfbb{R^p}$ cuando tenemos una distancia entre los elementos del conjunto original $\Omega$, tal que la distancia euclideana es la mejor aproximaci√≥n posible de las distancias originales.

Sea $(\Omega,\delta)$ el espacio $\Omega=\{\omega_1,...,\omega_n\}$ con medidas de disimilaridad (que tanto difieren) definidas. La matriz de distancias $\Delta$ se define como:

$$
\Delta = 
\left[ \begin{array}{ccccc}
\delta_{1 1} & \cdots & \delta_{1 n} \\
\vdots  & \ddots & \vdots \\
\delta_{n 1} & \cdots & \delta_{n n} \end{array} \right]
$$
lo anterior es la matriz de disimilaridades (dos a dos) de todos los lementos, en la diagonal hay puros ceros (por definici√≥n de disimilaridad), adem√°s es sim√©trica 

encontrar n vectores en R^p que preserven las distancias del conjunto original

Donde $\delta_{i j}=\delta(\omega_i, \omega_j)$, Dado MDS intenta preservar las distancias entonces$\delta_{i j}^2=d_E^2(\mathbf{x_i,x_j})$, es decir intentamos mapear vectores a Rp del conjunto omega original donde las distancias se "preservan"



delta^2 es cada elemento de la matriz de distancias al cuadrado
$$
\begin{align*}
\Delta^{(2)}&=
\left[ \begin{array}{ccccc}
\delta_{1 1}^2 & \cdots & \delta_{1 n}^2 \\
\vdots  & \ddots & \vdots \\
\delta_{n 1}^2 & \cdots & \delta_{n n}^2 \end{array} \right]\\
& =
\left[ \begin{array}{ccccc}
d_E^2(\mathbf{x_1,x_1}) & \cdots & d_E^2(\mathbf{x_1,x_n}) \\
\vdots  & \ddots & \vdots \\
d_E^2(\mathbf{x_n,x_1}) & \cdots & d_E^2(\mathbf{x_n,x_n}) \end{array} \right] \\ \\
& = \left[ \begin{array}{ccccc}
\mathbf{(x_1-x_1)'(x_1-x_1)} & \cdots & \mathbf{(x_n-x_1)'(x_n-x_1)}\\
\vdots  & \ddots & \vdots \\
\mathbf{(x_n-x_1)'(x_n-x_1)} & \cdots & \mathbf{(x_n-x_n)'(x_n-x_n)}
\end{array} \right] \\
& = \left[ \begin{array}{ccccc}
\mathbf{x_1'x_1+x_1'x_1-2x_1'x_1} & \cdots & \mathbf{x_1'x_1+x_n'x_n-2x_1'x_n}\\
\vdots  & \ddots & \vdots \\
\mathbf{x_n'x_n+x_1'x_1-2x_n'x_1} & \cdots & \mathbf{x_n'x_n+x_n'x_n-2x_n'x_n}
\end{array} \right] \\
& = \left[ \begin{array}{ccccc}
\mathbf{x_1'x_1} & \cdots & \mathbf{x_1'x_1}\\
\vdots  & \ddots & \vdots \\
\mathbf{x_n'x_n} & \cdots & \mathbf{x_n'x_n}
\end{array} \right] + 
\left[ \begin{array}{ccccc}
\mathbf{x_1'x_1} & \cdots & \mathbf{x_n'x_n}\\
\vdots  & \ddots & \vdots \\
\mathbf{x_1'x_1} & \cdots & \mathbf{x_n'x_n}
\end{array} \right] -2 
\left[ \begin{array}{ccccc}
\mathbf{x_1'x_1} & \cdots & \mathbf{x_1'x_n}\\
\vdots  & \ddots & \vdots \\
\mathbf{x_n'x_1} & \cdots & \mathbf{x_n'x_n}
\end{array} \right]  \\
& = \left[ \begin{array}{ccccc}
\mathbf{x_1'x_1} \\
\vdots \\
\mathbf{x_n'x_n}
\end{array} \right]\left[1 \cdots 1\right] + 
\left[ \begin{array}{ccccc}
1\\ \vdots \\ 1 \end{array} \right]
\left[\mathbf{x_1'x_1}  \cdots \mathbf{x_n'x_n}  \right] -2 
\left[ \begin{array}{ccccc}
x_{11} & \cdots & x_{1p}\\
\vdots  & \ddots & \vdots \\
x_{n1} & \cdots & x_{np}
\end{array} \right]
\left[ \begin{array}{ccccc}
x_{11} & \cdots & x_{n1}\\
\vdots  & \ddots & \vdots \\
x_{1p} & \cdots & x_{np}
\end{array} \right] \\
\Delta^{(2)} & = diag(\mathbf{XX'})(\mathbf{1_{n \times 1}})'+ \mathbf{1_{n \times 1}} diag(\mathbf{XX'})' -2 \mathbf{XX'} \\ \\
\end{align*}
$$

Sea la siguinete matriz:

$$
\mathbf{A} = -\frac{1}{2} \Delta^{(2)} \\
\mathbf{B} = (b_{ij}) = a_{ij}-a_{i.}-a_{.j}+a_{..}
$$

Donde $a_{i.} = \sum_{j=1}^n\frac{a_{ij}}{n} \quad a_{.j} = \sum_{i=1}^n\frac{a_{ij}}{n} \quad a_{..} =\sum_{i,j}^n\frac{a_{ij}}{n^2}$, Donde la matriz B puede ser escrita como: $\mathbf{B=HAH}$

La matriz de distancias $\Delta$ es una matriz de distancias euclideanas si y solo si $\mathbf{B} \geq 0$, i.e. los valores propios de $\mathbf{B}$ son no negativos.

**Demostraci√≥n**: La matriz de dsitancias $\Delta$ es una matriz de distancias euclideanas $\implies \mathbf{B} \geq 0$

$$
\begin{align*}
\mathbf{A}&=-\frac{1}{2}\mathbf{\left[diag(\mathbf{XX'})(\mathbf{1_{n \times 1}})'+ \mathbf{1_{n \times 1}}diag(\mathbf{XX'})'\right]+XX'}\\
\mathbf{HAH}&=\mathbf{H}\left[-\frac{1}{2}\mathbf{\left[diag(\mathbf{XX'})(\mathbf{1_{n \times 1}})'+ \mathbf{1_{n \times 1}}diag(\mathbf{XX'})'\right]+XX'}\right]\mathbf{H}\\
\mathbf{B}&=-\frac{1}{2}\mathbf{\left[Hdiag(\mathbf{XX'})(\mathbf{1_{n \times 1}})'\mathbf{H}+\mathbf{H}\mathbf{1_{n \times 1}}diag(\mathbf{XX'})'\mathbf{H}\right]+\mathbf{H}XX'\mathbf{H}}\\
\mathbf{B}&=-\frac{1}{2}\mathbf{\left[Hdiag(\mathbf{XX'})\mathbf{0_{n \times n}}+\mathbf{0_{n \times n}}diag(\mathbf{XX'})'\mathbf{H}\right]+\mathbf{H}XX'\mathbf{H}}\\
\mathbf{B}&=\mathbf{HXX'H}\\
\mathbf{B}&= \mathbf{HXX'H'} = \mathbf{HX(HX)ùêóùêó‚Ä≤'} \geq 0 \\
\end{align*}
$$

**Observaciones:**
Dado $\mathbf{A=XX'}$ entonces $b_{ii}=0-a_{i.}-a_{.i}+a_{..}$ y $b_{jj}=0-a_{j.}-a_{.j}+a_{..}$ entonces $\delta_{ij}=b_{ii}+b_{jj}-2b_{ij}=a_{ii}+a_{jj}-2a_{ij}$

**Demo**: $\mathbf{B} \geq 0 \implies $ la matriz de distancias $\Delta$ es una matroz de dsitancias euclideanas.

Dado $\mathbf{B} \geq 0 \implies $ entonces $\mathbf{B=YY'}$ Para alguna matriz $\mathbf{Y_{n \times p}}$ entonces $b_{ij}=\mathbf{y_i'y_i}$ donde $\mathbf{y_i'}$ su i-√©simo renglon $\mathbf{Y}$, y usando la ultima obssrvacion:

$\delta_{ij}^2=\mathbf{y_i'y_i}+\mathbf{y_j'y_j}-2\mathbf{y_i'y_j}=\mathbf{(y_i-y_j)'(y_i-y_j)}$

Entonces $\Delta$ es una matriz de distancias euclideanas.

DErivado de $\mathbf{B}$ es una matriz semidefinida positiva que permita una descomposici√≥n espectral a coordenadas compatibles con matriz de distancias.

$$
\begin{align*}
\mathbf{B} &= \mathbf{\Gamma \Lambda \Gamma'}\\
\mathbf{B} &= \mathbf{\Gamma \Lambda^{\frac{1}{2}} \Lambda^{\frac{1}{2}} \Gamma'} \\
\mathbf{B} &= \mathbf{\Gamma \Lambda^{\frac{1}{2}} (\Gamma \Lambda^{\frac{1}{2}})'} \\
\mathbf{B} &=
\left[ \begin{array}{ccccc}
\sqrt{\lambda_1} \mathbf{\gamma_1} \\
\vdots  \\
\sqrt{\lambda_n} \mathbf{\gamma_n}\end{array} \right]
\left[\sqrt{\lambda_1} \mathbf{\gamma_1} \dots \sqrt{\lambda_n} \mathbf{\gamma_n} \right]\\

\left[ \begin{array}{ccccc}

$$

### Escalamiento multidimensional no m√©trico
___

Sea $\Delta$  una matriz de distancias no euclideana. Existe un m√©todo que intenta convertir $\Delta$ a una matriz euclideana basada en la relacion de los elementos originales de $\Omega$ este metodo es llamado escalamiento multidimensional no m√©trico. El m√©todo de kruskal permite obtener una nueva matriz de distancias euclideanas basado en la minimizaci√≥n de una funci√≥n que mide la discrepancia entre las distancias originales y las distancias transformadas.

Este metodo considera que el orden de todas las diferentes distancias $\delta_{ij}$ es una propiedad asociada a $\Omega$.

$$\delta_{i_1j_1} \leq \delta_{i_2j_2} \leq \dots \leq \delta_{i_m j_m} \implies (\omega_{i_1},\omega_{j_1}) \leq (\omega_{i_2},\omega_{j_2}) \leq \dots \leq (\omega_{i_m},\omega_{j_m}) \in \Omega \times \Omega$$

Si las distancias transformadas $\delta'_{ij} = \varphi(\delta_{ij})$ se obtienen por $\varphi$ una funci√≥n monotona entonces el orden se preserva.

$$\delta_{i_1j_1} \leq \delta_{i_2j_2} \leq \dots \leq \delta_{i_m j_m} \implies \varphi(\delta_{i_1j_1}) \leq \varphi(\delta_{i_2j_2}) \leq \dots \leq \varphi(\delta_{i_m j_m})$$

Por lo tanto, las $\omega$s cercanas o lejanas segun $\delta_{i,j}$ ser√° tambi√©n cercano o lejano respecto a la distancia transformada $\delta'_{i,j}$, si adem√°s $\delta'_{i,j}$ la nueva matriz de distancias $\Delta'$ tiene soluci√≥n usando PCA.

Para la matriz no euclidena $\Delta$ considere ajustar las distancias obtenidas por PCA $d_{i,j}=d(\mathbf{x_i,x_j})$, dado $\Delta$ podr√≠a no ser descompuesto por el mapeo de PCA $\delta_{i_1,j_1} \rightarrow d_{i,j} $ no es monotona, la idea ahora es transformar $d_{i,j} \rightarrow d_{i,j}$ tal que $\delta_{i_1,j_1} \rightarrow d'_{i,j} $ es monotona. 

Para obtener la distancia transformada $d'$ el onjetivo es minimizar la funcion sobre todas las discrepancias de $d_{ij}$ y $\delta_{ij}$ i.e. $\sum_{i < j}(d_{ij}-\delta_{ij})^2$ sujeto a  $\delta_{i_1j_1} \leq \delta_{i_2j_2} \leq \dots \leq \delta_{i_m j_m}$

Una vez que tenemos nuestro nuevo conjunto de $\{ \delta'_{i_k j_k}\}_{k=0}^m$ es neceario asociar las coordenadas principales con la nueva matriz de distancias $\Delta' = (\delta'_{ij})$. 

PAra saber si la representaci√≥n obtenida refleja las distancias entre los nuevos vectores est√° bien, El SStress es calculado.

$$SStress = \sqrt{\frac{\sum_{i<j}(d'_{ij}-\delta_{ij})^2}{\sum_{i<j}(d'_{ij})^2}}$$

Corrije la posici√≥n de los puntos para optimizar el estr√©s de todos los puntos. La representaci√≥n se considera buena si SSstress no supera el 5%.

**Ventajas**
* Relaja las restricciones de distancias.
* Produce incrustaciones no lineales.

**Desventajas**
* Optimizaci√≥n iterativa altamente no lineal con m√≠nimos locales.
* No est√° claro c√≥mo elegir la transformaci√≥n de distancia.

```python
mds = manifold.MDS(n_components=2, max_iter=3000, eps=1e-9, random_state=seed,
                   dissimilarity="precomputed", n_jobs=1)

pos = mds.fit(similarities).embedding_

nmds = manifold.MDS(n_components=2, metric=False, max_iter=3000, eps=1e-12,
                    dissimilarity="precomputed", random_state=seed, n_jobs=1,
                    n_init=1)
npos = nmds.fit_transform(similarities, init=pos)
```

REescalanado y rotando los datos

```python
pos *= np.sqrt((X_true ** 2).sum()) / np.sqrt((pos ** 2).sum())
npos *= np.sqrt((X_true ** 2).sum()) / np.sqrt((npos ** 2).sum())
clf = skLearnPCA(n_components=2)
X_true = clf.fit_transform(X_true)
pos = clf.fit_transform(pos)
npos = clf.fit_transform(npos)
```

### **Ejemlplo 4: Distancia de Levenshtein**

La distancia de Levenshtein es una m√©trica de cadena para medir la diferencia entre dos cadenas. La distancia de Levenshtein entre dos palabras es el n√∫mero m√≠nimo de ediciones de un solo car√°cter (inserciones, eliminaciones o sustituciones) necesarias para cambiar una palabra por otra. Lleva el nombre del matem√°tico sovi√©tico Vladimir Levenshtein.

La distancia de Leveshtein entre dos cadenas${\displaystyle a,b}$  de largo $|a|,|b|$ respectivamente est√°n dadas por $\operatorname{lev}_{a,b}(|a|,|b|)$ donde:

$
\qquad\operatorname{lev}_{a,b}(i,j) = \begin{cases}
  \max(i,j) & \text{ si} \min(i,j)=0, \\
  \min \begin{cases}
          \operatorname{lev}_{a,b}(i-1,j) + 1 \\
          \operatorname{lev}_{a,b}(i,j-1) + 1 \\
          \operatorname{lev}_{a,b}(i-1,j-1) + 1_{(a_i \neq b_j)}
       \end{cases} & \text{ e.o.c.}
\end{cases}
$

```python
def levenshtein(str1, str2):
    '''
    esta funci√≥n calcula la distancia de 
    levenshtein entre dos cadenas
    '''
    d=dict()
    for i in range(len(str1)+1):
       d[i]=dict()
       d[i][0]=i
    for i in range(len(str2)+1):
       d[0][i] = i
    for i in range(1, len(str1)+1):
       for j in range(1, len(str2)+1):
          d[i][j] = min(d[i][j-1]+1, d[i-1][j]+1, d[i-1][j-1]+(not str1[i-1] == str2[j-1]))
    return d[len(str1)][len(str2)]
```

```python
words = ['looney','coney','money','donkey','honey','mark','park','dark','shark','spark']
#words = ['mono','tono','lomo','domo','cono','raiz','maiz','nariz','pais']

cols=['word1','word2','levenshtein']

wordsPD =  pd.DataFrame(words).assign(foo=1)
display(wordsPD)
cartesian = wordsPD.merge(wordsPD,on='foo').drop('foo', 1).values
distances = map(lambda x: [x[0],x[1],levenshtein(x[0],x[1])], cartesian)
display(distances)
display(cartesian)
distanceMatrixDF = pd.DataFrame(list(distances),columns=cols)\
                     .pivot(index='word1',columns='word2')

display(distanceMatrixDF)
```

```python
cols = ['x1','x2']

#escalamiento en R^2
mdsWordsVectors = manifold.MDS(n_components=2,
                               dissimilarity='precomputed',
                               random_state=5)\
                     .fit_transform(distanceMatrixDF.values)

mapWordsPD = pd.DataFrame(mdsWordsVectors,columns=cols)
mapWordsPD['word'] = list(distanceMatrixDF.reset_index()['word1'])

display(mapWordsPD[['word','x1','x2']])

plt.style.use('seaborn')
plt.figure(figsize=(15,7))
scatterPlot(plt,mapWordsPD,'x1','x2','darkblue',label='word')
```
