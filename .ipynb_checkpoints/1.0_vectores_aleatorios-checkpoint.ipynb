### Simulando una variable aleatoria continua
____

**Teorema de cambio de variable - corolario:**

Suponga que la variable aleatoria $X$ tiene de una función de distribución $F_X(x;\theta)$ si $Q \sim U(0,1)$ entonces la variable aleatoria $F_X^{-1}(Q;\theta) \sim X$

**ejemplo:**

Suponga que $X \sim exp(\lambda)$ entonces $F_X(x;\lambda)=1-e^{-\lambda x}$ entonces si q es un numero aleatorio entre [0,1] de la distribución $U(0,1)$ entonces x es una observación de $exp(\lambda)$:

hacemos q = f y despejamos x

$$x = -\frac{1}{\lambda}\ln(1-q)$$


```python
def gen_exp(lambda_param):
    '''
    generación de valores exponenciales
    
    regresa un valor de la distribución exponencial
    '''
    # generación de valores aleatorios 
    q = np.random.uniform(0,1)
    
    # función inversa de la exponencial
    return (-1/lambda_param)*np.log(1-q)

def gen_sample_expon(lambda_param, n):
    '''
    lambda param: es el parametro lambda de la exponencial 
    generación de vector de valores exponenciales
    de tamaño n
    
    regresa una lista
    '''
    #llenar un vector de valores exponenciales
    X = []
    for x in range(0,n):
        X.append(gen_exp(lambda_param))
    return X


#ejemplo de simulación de variable aleatorio exponencial
n = 500
lambda_param = 1/2
X = gen_sample_expon(lambda_param, n)
plt.figure(figsize=(15,8))
sns.histplot(X, bins=50, color='Blue')
```

```python
# comprobando la ley de los grandes numeros
# si aplicamos el limite cuando n tiende a infinito, 
# la media muestral converge a la media poblacional (teorica)


t = 5_000
estimators = []
lambda_param = 1/5


for i in range(1,t):
    X = gen_sample_expon(lambda_param, n=i)
    # promedio de los n valores exponenciales generados 
    estimators.append([i, np.mean(X)])
    
#se guardan los promedios de las iteraciones
df = pd.DataFrame(estimators, columns=['n','mean'])

# grafica de las medias conforme n (tamaño de la muestra aumenta)
plt.figure(figsize=(25,6))
plt.axhline(y=1/lambda_param, color='Red')
sns.lineplot(data=df, x='n', y='mean', color='Blue')
```

### Simualción de una variable aleatoria discreta com librerías de python

### simulación de una Gamma

```python
# generación de una variable aleatoria gamma

from scipy import stats 
# a =  alpha forma, b = beta escala
a, b = 3, 4
#
n = 1_000

#instancia de la gamma
gamma_dist = stats.gamma(a = a, scale = b)

# crear una muestra de la gamma rvs = random variable simulation
gamma_sample = gamma_dist.rvs(size = n)

# estimación de los parametros por maxima verosimilitud
gamma_dist_est = stats.gamma.fit(gamma_sample)


plt.figure(figsize=(15,8))
sns.histplot(gamma_sample, bins=50, color='Blue')

# lo que se imprime en medio es el parametro de localización
print('estimated',gamma_dist_est)
```

# Generación de vectores de vectores aleatorios
____

Sea el vector aleatorio $\mathbf{x}=(X_1,X_2)'$ con pdf (función de densidad):
$$ f_{X_1,X_2}(x_1,x_2) = 3x_1 \quad 0 \leq x_2 \leq x_1 \leq 1$$

tomando que:

$$ 
\begin{align*}
    f_{X_2|X_1}(x_1;x_2) &= \frac{f_{X_1,X_2}(x_1,x_2)}{f_{X_1}(x_1)}\\
    \implies f_{X_1,X_2}(x_1,x_2) &= f_{X_1}(x_1)f_{X_2|X_1}(x_2|x_1) \\
\end{align*}
$$

En este caso:


$$ 
\begin{align*}
    f_{X_1}(x_1) &= 3x_1^2 \quad 0 \leq x_2 \leq 1\\
    f_{X_2|X_1}(x_2;x_1) &= \frac{3x_1}{3x_1^2}\\
    f_{X_2|X_1}(x_2;x_1) &=\frac{1}{x_1} \quad 0 \leq x_2 \leq x_1\\
    \implies X_2 | X_1 &\sim U(0,x_1)
\end{align*}
$$

usando el corolario del teorema de cambio de variable para $q \sim U(0,1)$:
    
$$ 
\begin{align*}
    F_{X_1}(x_1) &= x_1^3 = q\\
    x_1 &= q^{1/3}
\end{align*}
$$

por último obtenemos:

$$ 
\begin{align*}
    x_1 &= q^{1/3} \quad \text{where } q \sim U(0,1)\\
    x_2 & \quad \text{where } X_2 | X_1 \sim U(0,x_1) \\
\end{align*}
$$
 
entonces el vector alatorio $\mathbf{x} = (x_1,x_2)$ tiene función de densidad $f_{X_1,X_2}(x_1,x_2)$

Este vector aleatorio tiene:
$$
\mathbf{\mu_x} = \begin{bmatrix} 
      0.75\\
      0.375\\
   \end{bmatrix} \quad
\mathbf{\Sigma_x} = \begin{bmatrix} 
      0.0375&0.01875\\
      0.1875&0.059375\\
   \end{bmatrix} \quad
\mathbf{\rho_x} = \begin{bmatrix} 
      1&0.397\\
      0.3973&1\\
   \end{bmatrix}
$$

```python
def sim_vector():
    # numero aleatorio entre 0,1
    q = np.random.uniform(0,1)
    #raiz cubica de 1
    x_1 = np.power(q, 1/3)
    # numero aleatorio uniforme entre 0 y x_1
    x_2 = np.random.uniform(0,x_1)
    #observacion
    return [x_1, x_2]

sim_vector()
```

Los numeros anteriores son una observación de: 
$$ f_{X_1,X_2}(x_1,x_2) = 3x_1 \quad 0 \leq x_2 \leq x_1 \leq 1$$

con correlación de ( matriz de correlaciones):
$$
\mathbf{\rho_x} = \begin{bmatrix} 
      1&0.397\\
      0.3973&1\\
   \end{bmatrix}
$$

generación de valores aleatorios del ejercicio anterior

```python
n = 5_000
X = []

#generamos n valores aleatorios provenientes de la distribución anterior
for i in range(0,n):
    X.append(sim_vector())

    #convertimos a data frame para poder plotear
data_matrix_df = pd.DataFrame(X, columns=['x_1','x_2'])
display(data_matrix_df.head(10))

plt.figure(figsize=(15,8))
sns.scatterplot(data=data_matrix_df, x='x_1', y='x_2', color='Blue')
```

recordemos el vector de medias

$$
\mathbf{\mu_x} = \begin{bmatrix} 
      0.75\\
      0.375\\
   \end{bmatrix} \quad
$$

```python
print(f'el vector de medias de la muestra aleatoria es \n{data_matrix_df.mean()}')
```

La matriz de covarianzas es
$$
\mathbf{\Sigma_x} = \begin{bmatrix} 
      0.0375&0.01875\\
      0.1875&0.059375\\
   \end{bmatrix} \quad
$$

```python
# matriz de varianzas y covarianzas
data_matrix_df.cov()
```

```python
fig = plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')
plt.rcParams['axes.facecolor']='white'

(X1, X2) = np.meshgrid(np.arange(0, 1, 0.02), np.arange(0, 1, 0.02))
Z = 3*X1
for i in range(0,len(Z)):
    for j in range(0,len(Z[0])):
        if j < i:
            Z[i][j] = np.nan
surf = ax.plot_surface(X1, X2, Z, cmap = 'Blues', alpha=0.8)
ax.scatter3D(data_matrix_df['x_1'], data_matrix_df['x_2'], [0]*len(data_matrix_df), color = "blue")
ax.set_xlabel('X1', labelpad=20)
ax.set_ylabel('X2', labelpad=20)
ax.set_zlabel('$f(X_1,X_2)$', labelpad=20)
ax.view_init(30, 100)
plt.show()
```

```python
cov_matrix_df = pd.DataFrame([[0.0375,0.01875],[0.01875,0.05937]], columns=['x_1','x_2'],index=['x_1','x_2'])
plt.figure(figsize=(4,3))
plt.title('Population covariance')
sns.heatmap(data=cov_matrix_df, annot=True,fmt=".4f",cmap="Blues")

plt.figure(figsize=(4,3))
plt.title('Sample covariance')
sns.heatmap(data=data_matrix_df.cov(), annot=True,fmt=".4f",cmap="Blues")
```

```python
corr_matrix_df = pd.DataFrame([[1,0.397],[0.397,1]], columns=['x_1','x_2'],index=['x_1','x_2'])

plt.figure(figsize=(4,3))
plt.title('Population correlation')
sns.heatmap(data=corr_matrix_df, annot=True,fmt=".4f",cmap="Blues")

plt.figure(figsize=(4,3))
plt.title('Sample correlation')
sns.heatmap(data=data_matrix_df.corr(), annot=True,fmt=".4f",cmap="Blues")
```
