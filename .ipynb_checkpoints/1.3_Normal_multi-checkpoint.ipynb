# 2.1 Normal multivariada


Denotamos el vector aleatorio $\textbf{x}$ que distribuye normal multivariado con parametros $\mu \in \mathbb{R}^p$ y matriz semidefinida positiva $\Sigma \in \mathbb{R}^p \times \mathbb{R}^p$ como:
$$
\textbf{x} \sim N_p\left(\mu , \Sigma \right) 
$$

si $\textbf{x}$ tiene distribuciones dadas por:

$$
f_{X_1,...,X_p}(x_1,...,x_p) = \frac{\left| \Sigma \right|^{-\frac{1}{2}}}{\left[\sqrt{2 \pi}\right]^p} exp\left( -\frac{1}{2} \left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right] \right)
$$

Esta distribución Multivariada es llamda la **Distribución Normal Multivariada**. Nota que la función de densidad del vector aleatorio es completamente definido por los parametros $\mu$ y $\Sigma$, donde $\Sigma$ debería de ser simétrica con determinante distinto de cero.

cuando $\mu=\textbf{0}$ and $\Sigma=\textbf{I}$ entonces se dice que $\mathbf{x}$ distribuye como una normal multivariada estándar.

el argumento $\left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right]$ es una distancia ponderada al cuadrado , cuando $\Sigma = \mathbf{I}$ then $\left[\textbf{x}-\mu \right]' \Sigma^{-1} \left[\textbf{x} - \mu \right] = \left|\left|\mathbf{x}-\mu\right|\right|^2$.

#### Propiedades

Si $\mathbf{x} \in \mathbb{R}^p, \mathbf{x} \sim N\left(\mu , \Sigma \right)$ y
1. $\mathbf{A} \in \mathbb{R}^p \times \mathbb{R}^q$ matriz de constantes, entonces $\mathbf{x_{1 \times p}A_{p \times q}} \sim \mathbf{N_q( A \mu, A \Sigma A')}$.
1. $\mathbf{\Sigma=\Sigma^{1/2} \Sigma^{1/2}}$ donde $\Sigma^{1/2}\Sigma^{1/2}$ es la descomposición en **Factorización de cholesky** de $\mathbf{\Sigma}$ entonces $\mathbf{z}=\mathbf{\left( x -\mu \right)\Sigma^{-1/2}}\sim N_p\left(\mathbf{0}, \mathbf{I} \right)$

#### Distribución normal bivariada
___

Llamamos a la distribución conjunta anterior para el vector $\textbf{x}=(X_1,X_2)$ la distribución nromal bivariada:


$$
\begin{align*}
f_{X_1,X_2}(x_1,x_2) & = \frac{1}{2 \pi (1-\rho^2)} exp\left( -
\frac{1}{2 (1-\rho^2)}
\left[ \left(\frac{x_1-\mu_1}{\sigma_{1}}\right)^2+
\left(\frac{x_2-\mu_2}{\sigma_{2}}\right)^2-
2\rho \frac{(x_1-\mu_1)(x_2-\mu_2)}{\sigma_1 \sigma_2}
\right]
\right)\\
\end{align*}
$$

**Example 2.1 - 1 [Visualización de una densidad normal bivariada]**

$$
f_{X_1,X_2}(x_1,x_2), 
(X_1,X_2) \sim N\left(
\mu=\left[ \begin{array}{ccccc}
0 \\
0 \end{array} \right]
,
\Sigma=\left[ \begin{array}{ccccc}
2 & 1\\
1 & 2 \end{array} \right]
\right) 
$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

mu=[0, 0] # vector de medias
sigma = [[2,1],[1,2]] #matriz de covarianzas

plt.rcParams['axes.facecolor'] = 'white' # grafico en blanco

y = np.linspace(-4,4,500) # ejes
x = np.linspace(-4,4,500) # ejes

X, Y = np.meshgrid(x,y) # maya de puntos de todos los pares ordenados


pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y

# normal multivario de la paquería sci stats
rv = multivariate_normal(mu,sigma) # variable aleatoria

fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
# rv.pdf (funcion de densidad de probabilidad) evaluada en pos
ax.plot_surface(X, Y, rv.pdf(pos),cmap='plasma',linewidth=0)

#nombres de ejes
ax.set_xlabel('$X_1$ axis')
ax.set_ylabel('$X_2$ axis')
ax.set_zlabel('$f(x_1,x_2)$')
ax.view_init(50,120) #angulo de vision del gráfico
plt.show()
```

# Descomposición espectral
___


```python
import numpy as np

S = np.matrix([[2,1],[1,2]])

#matriz que tiene a los vectores propios como columnas
Gamma = np.matrix([[1/np.sqrt(2),-1/np.sqrt(2)],[1/np.sqrt(2),1/np.sqrt(2)]])

#matriz de valores propios
Lambda = np.matrix([[3,0],[0,1]])

#como es semidefinida positiva, la inversa coincide con la traspuesta
GammaInv = Gamma.transpose() 

print('Matrix S\n',S)
print('Matrix Gamma\n',Gamma)
print('Matrix Lambda\n',Lambda)
print('Matrix GammaInv\n',GammaInv)

# recuperación de la matriz sigma de gamma con lamda y eso con gamma traspuesta
### descomposición espectral
print('Gamma*Lambda*GammaInv \n',np.dot(Gamma,np.dot(Lambda,Gamma.transpose())))
```

**steep 2** codificación de funciones para realizar el análisis de contornos de densidad de probabilidad de constantes

* **computeMaximumLikelihoodEstimators**: calcular el vector de medias de la muestra $\mathbf{\bar{x}}$y la matriz de covarianza de la muestra $\mathbf{S}$ ya que estos estadísticos son los estimadores de máxima verosimilitud de los parámetros de distribución normal multivariante.

* **diagonalize**: Realiza la descomposición espectral de la matriz de covarianza de la muestra, debido a que la descomposición espectral es función de $\mathbf{\Sigma}$ por el teorema de la invariancia, los valores propios y los vectores propios que se obtienen al descomponer la matriz de covarianza de la muestra serán los estimadores de los vectores propios y los valores propios poblacionales.

* **plotConfidenceEllipse**: Una vez que se da el valor crítico de chi cuadrado, es hora de trazar el contorno de densidad de probabilidad constante en la dirección de los vectores propios con el valor propio de longitud multiplicado por el valor crítico de chi cuadrado

* **getProbabilityDensityContour**: Ordena la última función y realiza el análisis.

```python
from matplotlib.patches import Ellipse
from scipy.stats import chi2

def computeMaximumLikelihoodEstimators(datasetPD,columns):
    '''
    recibe un data frame y las columnas de las cuales quiero sacar
    el vector de medias y matriz de cob¿variazas
    
    regresa una lista-vector de medias
    y una matriz de cvov
    '''
    mu = list(datasetPD[columns].mean())
    sigma = np.matrix(datasetPD[columns].cov())
    return mu, sigma

def diagonalize(sigma):
    '''
    recibe una matriz sigma de varianzas y cov
    y lo unico que se hace es la diagonalización de 
    la matriz
    
    está matriz regresa los valores y vectores propios
    
    los valores propios los ordeno y en el orden de los valores propios
    me traigo los vectores propios (en ese orden)
    
    '''
    eigenValues, eigenVectors = np.linalg.eigh(sigma) # sacar valores y vectores propios
    idx = eigenValues.argsort()[::-1]    # los ordeno de mayor a menor
    eigenValues = eigenValues[idx] # guardo los valores propios ordenados 
    eigenVectors = eigenVectors[:,idx] # guardo los vectores propios ordenados
    return eigenValues, eigenVectors # regresa valores y vectores propios

def plotConfidenceEllipse(plt,mu,eigenValues,eigenVectors,chiSquaredCriticalVale,ax=None,**kwargs):
    '''
    graficar para datos en dos dimensiones, porque nomás el arco tangente existe en 2 dimensiones
    
    
    
    '''
    
    if ax is None:
        ax = plt.gca()
    
    #en python ya se pueden graficar las elipses
    
    #sacar el angulo con el arcotangente de los valores propios para tener el grado de inclinación de los 
    # vectores propios
    theta = np.degrees(np.arctan2(*eigenVectors[:,0][::-1]))
    
    #largo y ancho provenientes de la chi cuadrada
    width,height = chiSquaredCriticalVale * np.sqrt(eigenValues)
    
    #elipse de confianza
    confidenceEllipse = Ellipse(xy=mu,width=width,height=height,angle=theta,**kwargs)
    
    # funcion para plotear
    ax.add_artist(confidenceEllipse)
    return confidenceEllipse

def getProbabilityDensityContour(plt,datasetPD,columns,alpha,freedomDegrees,color,name='DataSet'):
    '''
    esta funcion manda a llamar todo lo anterior para realizar para hacerlo en orden 
    
    '''
    #obtener mu y sigma
    mu, sigma = computeMaximumLikelihoodEstimators(datasetPD,columns)
    
    #eigenvalores
    eigenValues, eigenVectors = diagonalize(sigma)
    
    #valor critico de la chi
    chiSquaredCriticalVale = chi2.ppf(q=(1-alpha),df=freedomDegrees)
    
    #grafica
    plotConfidenceEllipse(plt,mu,eigenValues,eigenVectors,chiSquaredCriticalVale,color=color,alpha=0.1)
    
    #tabal resumen
    sumaryTable=[]
    sumaryTable.append(['Mean',mu])
    sumaryTable.append(['Covariance matrix',sigma])
    sumaryTable.append(['EigenValues',eigenValues])
    sumaryTable.append(['EigenVectors',eigenVectors])
    sumaryTable.append(['Confidence',1-alpha])
    sumaryTable.append(['Chi-squared critical value',chiSquaredCriticalVale])
    sumaryTableDF = pd.DataFrame(sumaryTable,columns=['Summary',name])
    return sumaryTableDF

def scatterPlot(plt,datasetPD,col_1,col_2,color):
    '''
    funcion scatter de siempre
    '''
    dataPD = datasetPD[[col_1,col_2]]
    plt.scatter(dataPD[col_1],dataPD[col_2],color=color)
```

**Example 2.1 - 3 
[Generación de una muestra de distribución normal multivariante para p dimensiones]**

```python
import numpy as np
def getMvNormalSample(mu,sigma,n):
    '''
    Generación de una normal mulltivariada 
    dadas miu y sigma, la n indica el numero de muestras a obtener
    
    '''
    
    columns = ['x'+str(i+1) for i in range(0,len(mu))]
    
    #generación de una normal multivarriada con los parametros dados
    return pd.DataFrame(np.random.multivariate_normal(mu,sigma,n),columns=columns)
```

$$
