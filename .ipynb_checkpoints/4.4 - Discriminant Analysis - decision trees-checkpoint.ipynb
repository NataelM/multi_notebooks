Decision tree learning is a method commonly used in data mining. The goal is to create a model that predicts the value of a target variable based on several input variables.

A decision tree is a simple representation for classifying examples. For this section, assume that all of the input features have finite discrete domains, and there is a single target feature called the "classification". Each element of the domain of the classification is called a class. A decision tree or a classification tree is a tree in which each internal (non-leaf) node is labeled with an input feature.

<img src='static/dtree.png' width="80%" height="80%">

Algorithms for constructing decision trees usually work top-down, by choosing a variable at each step that best splits the set of items. Different algorithms use different metrics for measuring "best". These generally measure the homogeneity of the target variable within the subsets. Some examples are given below. These metrics are applied to each candidate subset, and the resulting values are combined (e.g., averaged) to provide a measure of the quality of the split.


The process of growing a decision tree can be expressed as a recursive algorithm as follows:

1) Pick the feature that when parent node is split, it results in the largest information gain.

2) Stop if child nodes are pure or no improvement in class purity can be made.

3) Go back to step 1 for each of the two child nodes.

There are many measures that can be used to determine the best way to split the records. These measures are defined in terms of the class distribution of the records before and after splitting. Let $p(i|t)$ denote the fraction of records belonging to class i at a given node t. We sometimes omit the reference to node t and express the fraction as $p_i$ . In a two-class problem, the class distribution at any node can be written as $(p_0 , p_1)$.

The measures developed for selecting the best split are often based on the degree of impurity of the child nodes. The smaller the degree of impurity, the more skewed the class distribution. For example, a node with class distribution (0, 1) has zero impurity, whereas a node with uniform class distribution (0.5, 0.5) has the highest impurity.

```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

def compute_frequencies(dataset_df,col,index=None):
    if index is None:
        index = 'index'
    columns=[col,'freqs_'+col,'relative_freqs_'+col]
    freqs_df = dataset_df.reset_index()\
                .groupby(col)\
                .count()\
                .reset_index()[[index,col]]\
                .rename(columns={index:'freqs_'+col})
    
    freqs_df['relative_freqs_'+col] = freqs_df['freqs_'+col]/len(loans_raw_df)
    freqs_df= freqs_df[columns].sort_values(col)
    missings_counts = loans_raw_df[col].isna().sum()
    missings_df = pd.DataFrame([['Missings',missings_counts,missings_counts/len(loans_raw_df)]],columns=columns)
    freqs_df = pd.concat([freqs_df, missings_df])
    return freqs_df
```

### Iris Dataset
___

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree

n_classes = 3
plot_colors = "ryb"
plot_step = 0.02
iris = load_iris()
plt.figure(figsize=(13,10))
for pairidx, pair in enumerate([[0, 1], [0, 2], [0, 3],[1, 2], [1, 3], [2, 3]]):
    X = iris.data[:, pair]
    y = iris.target
    clf = DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=5).fit(X, y)
    plt.subplot(2, 3, pairidx + 1)
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu)
    plt.xlabel(iris.feature_names[pair[0]])
    plt.ylabel(iris.feature_names[pair[1]])
    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c=color, label=iris.target_names[i],
                    cmap=plt.cm.RdYlBu, edgecolor='black', s=15)

plt.legend(loc='lower right', borderpad=0, handletextpad=0)
plt.axis("tight")
plt.figure(figsize=(15,10))
clf = DecisionTreeClassifier().fit(iris.data, iris.target)
```

```python
import graphviz 
from sklearn import tree
dot_data = tree.export_graphviz(clf, out_file=None, 
feature_names=iris.feature_names,  
class_names=iris.target_names,  
filled=True, rounded=True,  
special_characters=True)  
graph = graphviz.Source(dot_data)  
graph
```

### Loans Dataset
___

```python
loans_raw_df = pd.read_csv('DataSets/loans_prediction.csv')
loans_raw_df
```

```python
categoricals = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Credit_History']
map_dictionary = {}
for col in categoricals:
    freqs_df = compute_frequencies(loans_raw_df,col,index='Loan_ID')
    display(freqs_df)
    colum_values = list(freqs_df[col])[:len(freqs_df)-1]
    column_dict = {}
    for i,value in enumerate(colum_values):
        column_dict.update({value:str(int(i))})
    column_dict.update({np.nan:np.nan})
    map_dictionary.update({col:column_dict})
loans_cleaned_df = loans_raw_df.replace(map_dictionary).copy()
loans_cleaned_df
```

```python
numericals = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']
loans_cleaned_df = loans_raw_df.replace(map_dictionary).copy()

for col in numericals:
    loans_cleaned_df[col] = loans_raw_df.apply(lambda row: float(row[col]),axis=1)

loans_cleaned_df['IndividualIncome'] = (loans_cleaned_df['ApplicantIncome']+loans_cleaned_df['CoapplicantIncome'])/(loans_cleaned_df['Dependents']+1)
loans_cleaned_df['Installment'] = loans_cleaned_df['LoanAmount']/loans_cleaned_df['Loan_Amount_Term']

feature_engineering = ['IndividualIncome','Installment']
numericals+=feature_engineering
loans_cleaned_df
```

```python
for col in numericals:
    loans_cleaned_df[col] = loans_cleaned_df.apply(lambda row: np.log(float(row[col])+1),axis=1)
    
loans_cleaned_df.set_index('Loan_ID',inplace=True)
loans_cleaned_df.rename(columns={'Loan_Status':'target'},inplace=True)
loans_cleaned_df.replace({'target':{'Y':0,'N':1}},inplace=True)
loans_cleaned_df
```

```python
smp = sns.pairplot(loans_cleaned_df[numericals+['target']],hue="target",diag_kind='hist',palette="plasma")
```

```python
loans_train_df, loans_test_df = train_test_split(loans_cleaned_df, test_size=0.4, random_state=42)
loans_train_df.head(10)
```

```python
from sklearn.impute import SimpleImputer
imputer_cat = SimpleImputer(missing_values=np.nan, strategy='most_frequent').fit(loans_train_df[categoricals])
impuer_cont = SimpleImputer(missing_values=np.nan, strategy='mean').fit(loans_train_df[numericals])

imputed_train_cat_X = imputer_cat.transform(loans_train_df[categoricals])
imputed_train_cont_X = impuer_cont.transform(loans_train_df[numericals])

imputed_train_cat_df = pd.DataFrame(imputed_train_cat_X,columns=categoricals,index=loans_train_df.index)
imputed_train_cont_df = pd.DataFrame(imputed_train_cont_X,columns=numericals,index=loans_train_df.index)
imputed_train_df = imputed_train_cat_df.merge(imputed_train_cont_df,left_index=True, right_index=True)
imputed_train_df['target'] = loans_train_df['target']
imputed_train_df
```

```python
from sklearn.preprocessing import OneHotEncoder

one_hot_encoder = OneHotEncoder(sparse=False).fit(imputed_train_df[categoricals])
dummy_matrix_train_df = pd.DataFrame(one_hot_encoder.transform(imputed_train_df[categoricals]),
                                      columns=list(one_hot_encoder.get_feature_names(categoricals)))
dummy_columns = list(one_hot_encoder.get_feature_names(categoricals))
dummy_matrix_train_df['Loan_ID'] = imputed_train_df.index
dummy_matrix_train_df.set_index('Loan_ID',inplace=True)
dummy_matrix_train_df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA 

kw = 6
standard_scaler = StandardScaler().fit(imputed_train_df[numericals])
std_data_matrix_train = standard_scaler.transform(imputed_train_df[numericals])
pca_model = PCA(n_components=kw).fit(std_data_matrix_train)
train_pca_X = pca_model.transform(std_data_matrix_train)
components_cols = ['pc_'+str(i+1) for i in range(0,kw)]

pca_train_df = pd.DataFrame(train_pca_X,columns=components_cols)
pca_train_df['Loan_ID'] = imputed_train_df.index
pca_train_df.set_index('Loan_ID',inplace=True)
pca_train_df
```

### xgBoost
___

```python
from xgboost import XGBClassifier

design_matrix_train_df = dummy_matrix_train_df.merge(pca_train_df,left_index=True, right_index=True)
target_train_df = imputed_train_df[['target']].copy()

dtree_clf = XGBClassifier(max_depth=2,
                learning_rate=0.1,
                gamma=1,
                subsample=0.5,
                scale_pos_weight=0.5,
                random_state=45)

dtree_clf.fit(design_matrix_train_df,target_train_df[['target']])

target_train_df['score'] = dtree_clf.predict_proba(design_matrix_train_df)[:,1]
target_train_df
```

```python
imputed_test_cat_X = imputer_cat.transform(loans_test_df[categoricals])
imputed_test_cont_X = impuer_cont.transform(loans_test_df[numericals])

imputed_test_cat_df = pd.DataFrame(imputed_test_cat_X,columns=categoricals,index=loans_test_df.index)
imputed_test_cont_df = pd.DataFrame(imputed_test_cont_X,columns=numericals,index=loans_test_df.index)

imputed_test_df = imputed_test_cat_df.merge(imputed_test_cont_df,left_index=True, right_index=True)
imputed_test_df['target'] = loans_test_df['target']
imputed_test_df.head(10)

dummy_matrix_test_df = pd.DataFrame(one_hot_encoder.transform(imputed_test_df[categoricals]),
                                     columns=list(one_hot_encoder.get_feature_names(categoricals)))

dummy_matrix_test_df['Loan_ID'] = imputed_test_df.index
dummy_matrix_test_df.set_index('Loan_ID',inplace=True)


std_data_matrix_test = standard_scaler.transform(imputed_test_df[numericals])
test_pca_X = pca_model.transform(std_data_matrix_test)

pca_test_df = pd.DataFrame(test_pca_X,columns=components_cols)
pca_test_df['Loan_ID'] = imputed_test_df.index
pca_test_df.set_index('Loan_ID',inplace=True)

design_matrix_test_df = dummy_matrix_test_df.merge(pca_test_df,left_index=True, right_index=True)
target_test_df = imputed_test_df[['target']].copy()
target_test_df['score'] = dtree_clf.predict_proba(design_matrix_test_df)[:,1]
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

train_fpr, train_tpr, train_thresholds  = roc_curve(target_train_df['target'],target_train_df['score'])
test_fpr, test_tpr, test_thresholds  = roc_curve(target_test_df['target'],target_test_df['score'])

plt.figure(figsize=(10,10))
plt.plot(train_fpr,train_tpr, linestyle='-', color='blue', label='Training ROC')
plt.plot(test_fpr,test_tpr, linestyle='-', color='red', label='Testing ROC')
plt.plot([0,1],[0,1], color='black', label='Randomness Base Line')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

print('AUC TRAIN:',roc_auc_score(target_train_df['target'],target_train_df['score']))
print('AUC TEST:',roc_auc_score(target_test_df['target'],target_test_df['score']))
print('GINI TRAIN:',2*roc_auc_score(target_train_df['target'],target_train_df['score'])-1)
print('GINI TEST:',2*roc_auc_score(target_test_df['target'],target_test_df['score'])-1)
```
