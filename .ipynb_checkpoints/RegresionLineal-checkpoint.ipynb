```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
from sklearn.decomposition import PCA
from IPython.display import display, HTML
warnings.filterwarnings('ignore')
pd.set_option('max_colwidth',100)
plt.style.use('ggplot')

def clean_and_cast(value):
    value = str(value)
    cleaned_value = [char for char in value if char!=","]
    try:
        return float(''.join(cleaned_value))
    except:
        return None
```

Carga y limpieza del dataset

```python
numerical = ['confirmado_acumulados','casos_por_millon','recuperadas','muertes']
covid_df = pd.read_csv('DataSets/covid19.csv')
covid_df.set_index('Ubicación',inplace=True)

for col in covid_df:
    covid_df[col] = covid_df[col].apply(lambda row: clean_and_cast(row))
covid_df
```

Queremos predecir el número de pacientes recuperados por pais en funcion de los confirmados acumulados, casos por millon y muertes para poder estimar el número de recuperados en aquellos paises que tienen este dato perdido. Se ajusta un modelo de regresion lineal multiple.

```python
import statsmodels.api as sm

y = 'recuperadas'
x = ['confirmado_acumulados','casos_por_millon','muertes']

Xreg = sm.add_constant(covid_df[~covid_df[y].isnull()][x])
Yreg = covid_df[~covid_df[y].isnull()][y]
Xymissings = sm.add_constant(covid_df[covid_df[y].isnull()][x],prepend=True)

linear_regresion_model = sm.OLS(Yreg, Xreg, prepend=True).fit()
display(linear_regresion_model.summary())

plt.figure(figsize=(5,5))
ax = sns.scatterplot(x=Yreg,y=linear_regresion_model.predict(Xreg),color='Blue')
```

Con el modelo ajustado se imputan las entradas perdidas en la columna de recuperados.

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

non_missings_recovery = covid_df[~covid_df[y].isnull()]
missings_recovery = covid_df[covid_df[y].isnull()]    
missings_recovery['recuperadas'] = linear_regresion_model.predict(sm.add_constant(missings_recovery[x],prepend=True))
covid_19_imputed_df = pd.concat([non_missings_recovery,missings_recovery])
covid_19_imputed_df
```
