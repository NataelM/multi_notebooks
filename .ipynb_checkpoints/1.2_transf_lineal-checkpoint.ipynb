# Transfromaciones lineales sobre la matriz de datos

**Ejemplo 1.2 - 1 [Calculos de transformaciones lineales]**

```python
import pandas as pd

gradesPath = 'DataSets/Grades.txt'
columns = ['Phisics','Maths','French','History','Literature']
gradesPD = pd.read_csv(gradesPath)
gradesPD
```

```python
import numpy as np
np.matrix(gradesPD[columns].values)
np.matrix([[1/5, 1/5, 1/5, 1/5, 1/5], # un quinto porque al hacer el producto matricial dará el promedio de todas las materias(5)
         [1/2, 1/2, 0, 0, 0], # un medio para sacar el promedio de las materias de ciencias(2)
         [0, 0, 1/3, 1/3,1 /3]]).transpose() # un tercio apra sacar el promedio de las materias de humanisticas (3)
```

```python
import numpy as np
gradesM = np.matrix(gradesPD[columns].values)
matrixT = np.matrix([[1/5,1/5,1/5,1/5,1/5],
                     [1/2,1/2,0,0,0],
                     [0,0,1/3,1/3,1/3]]).transpose()

#producto mastricial para obtener
## el promedio general
## el promedio de las humanisticas
## el promedio de las de ciencias 
transformedDataMatrix = np.dot(gradesM,matrixT)
print('Transformation matrix\n',matrixT)

pd.DataFrame(transformedDataMatrix, columns = ['OverallMean', 'ScienceMean',' LettersMean'])
```

**Example 1.2 - 1 [Estandarización de las calificaciones promedio DF y cálculo de las medidas de variabilidad general]**

Estandarizar las calificaciones promedio de un subconjunto de columnas (Física, Matemáticas, Literatura) evitando la unión cruzada
___

```python
def getStdDataMatrix(data, columns, keys):
    '''
    Entendemos porestandarización a restar la media 
    y dividir entre la varianza
    data: df 
    columns: columnas a estandarizar
    keys: columnad que no voy a centrar/estandarizar
    
    regresa un df con las columnas estandarizada
    '''
    dataframePD = data.copy(deep = True)
    
    #vecctor de medias en diccionario
    meanVector = dataframePD.mean().to_dict()
    
    #vector de desviaciones estandar en diccionario
    stdVector = dataframePD.std().to_dict()
    
    # por columna estandarizar 
    for col in columns:
        dataframePD['std_'+col] = (dataframePD[col]-meanVector[col])/stdVector[col]
    return dataframePD[keys+['std_'+col for col in columns]]
```

```python
columns = ['Phisics', 'Maths', 'Literature']
stdGradesPD = getStdDataMatrix(gradesPD, columns, keys=['Student'])
stdGradesPD
```

Lo anterior es una matriz de datos, y la matriz de covarianza de los datos estandarizados coincide con la matriz de correlaciones de los datos sin estandarizar

```python
# cov estandarizada
covarianceSz =  stdGradesPD[['std_'+col for col in columns]].cov()
covarianceSz
```

```python
#corr de los datos no estandarizasa
correlationRx = gradesPD[columns].corr()
correlationRx
```
